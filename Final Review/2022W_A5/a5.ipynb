Find the team name that has the highest average HomeScore as the home team.

```python
home_scores = games.merge(teams, left_on="HomeTeamID", right_on="TeamID", how="inner")[["TeamName", "HomeScore"]]
home_scores
```

```python
avg_table = home_scores.groupby('TeamName')['HomeScore'].mean().reset_index()
avg_table.head(5)
```

```python
result = avg_table.sort_values(by="HomeScore", ascending=False).head(1)['TeamName']
result
```

```python
highest_home_score = avg_table['HomeScore'].max()
highest_home_score
```

```python
result = avg_table[avg_table['HomeScore'] == highest_home_score].iloc[0, 0]
result
```

Which FieldType has the highest total scores (HomeScore+AwayScore)? Which has the lowest? (Extra, ungraded) 


```python
games['TotalScore'] = games['HomeScore'] + games['AwayScore']
games.head(5)
```

```python
games['FieldType'] = games['FieldType'].str.replace(' ', '', regex=False)

field_total = games.groupby('FieldType')['TotalScore'].sum().reset_index()
field_total
```

```python
field_total[field_total['TotalScore'] == field_total['TotalScore'].max()]['FieldType']
```
