iii) Add a column that calculates the return of that day. Assume that the daily return is computed using (Current_Close-Previous_Close)/Previous_Close

```python
df['PrevClose'] = df['Close'].shift()

# 把 row 0 的 NaN 填上
df.loc[0, 'PrevClose'] = df.loc[0, 'Close']

df['Return'] = (df['Close'] - df['PrevClose']) / df['PrevClose']

df.drop(['PrevClose'], inplace=True, axis=1)

df
```
