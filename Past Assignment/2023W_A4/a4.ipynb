```python
import pandas as pd

df = pd.read_csv('Movies.csv')

df.fillna(0, inplace=True)
df.head()
```

1. Create a dataframe that counts the number of movies by genre and distributor. Calculate the total
 number of movies.

```python
genre_distributor_count_df = df.groupby(['Genre', 'Distributor']).size().reset_index(name='NumberOfMovies')
genre_distributor_count_df
```

```python
genre_distributor_count_df['NumberOfMovies'].sum()
```

2. Which distributor distributed the highest percentage of action movies?  
Which distributor
 distributed the highest percentage of adventure movies?

```python
action_df = genre_distributor_count_df[genre_distributor_count_df['Genre'] == 'Action']
action_df[action_df['NumberOfMovies'] == action_df['NumberOfMovies'].max()]['Distributor']

```

3. For each distributor (i.e., there are a total of 28 distributors) show the total number of movies distributed by genre.  
 What is the total number of movies distributed by each distributor?  
 What is the total number of movies by genre?

```python
distributor_genre_count_df = df.groupby(['Distributor', 'Genre']).size().reset_index(name='NumberOfMovies')
distributor_genre_count_df
```

```python
distributor_count_df = df.groupby(['Distributor']).size().reset_index(name='NumberOfMovies')
distributor_count_df
```

```python
genre_count_df = df.groupby(['Genre']).size().reset_index(name='NumberOfMovies')
genre_count_df
```

4. For Buena Vista, which genre represented the highest percentage of genres distributed?

```python
vista_df = distributor_genre_count_df[distributor_genre_count_df['Distributor'] == 'Buena Vista']
vista_df[vista_df['NumberOfMovies'] == vista_df['NumberOfMovies'].max()]['Genre']
```

5. What is the average US Gross sales for all movies for each distributor?  
What is the average US Gross sales for all movies for each genre?  
 What is the average US Gross sales of all movies?

```python
# 1)
df.groupby('Distributor')['Total_US_Gross'].mean().reset_index(name='Average_US_Gross')

# 2)
df.groupby('Genre')['Total_US_Gross'].mean().reset_index(name='Average_US_Gross')

# 3)
df['Total_US_Gross'].mean()
```
